# Random Variables
## Definition
A random variable is a variable whose possible values are numerical outcomes of a random phenomenon. There are two types: discrete and continuous.

### Example
`Discrete Random Variable`: The number of heads in 10 coin flips.


`Continuous Random Variable`: The height of students in a class.

# Python Example

```python
import numpy as np

# Discrete Random Variable: Number of heads in 10 coin flips

coin_flips = np.random.binomial(n=10, p=0.5, size=1000)

print("Discrete",coin_flips[:10])


```

```python
import matplotlib.pyplot as plt
plt.hist(coin_flips)
plt.show()
```

```python
# Continuous Random Variable: Height of students



heights = np.random.normal(loc=170, 
                           scale=10, 
                           size=1000)

print("Continuous", heights[:10])
```

```python
import matplotlib.pyplot as plt
plt.hist(heights)

plt.show()
```

# Descriptive Statistics
### Definition
Descriptive statistics summarize or describe the characteristics of a data set. They include measures like mean, median, mode, range, variance, and standard deviation.

##### Example
`Example`: Summarizing the test scores of students to find the average score.

### Mathematical Intuition

# 

# Python Example

```python
import numpy as np
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

scores = [85, 90, 78, 92, 88, 
          76, 95, 89, 84, 91]

mean = np.mean(scores)
median = np.median(scores)
mode = stats.mode(scores)

print(f'''Mean: {mean}, 
Median: {median}, 
Mode: {tuple(mode)[0]}''')
```

# Measure of Central Tendency
### Definition
Measures of central tendency are values that describe the center of a data set. The main measures are mean, median, and mode.


#### Example
Finding the average number of hours people spend on their phones daily.

# Python Example

```python
usage_times = [2, 3, 3, 4, 4, 4, 5, 5, 6, 7]

mean = np.mean(usage_times)
median = np.median(usage_times)
mode = stats.mode(usage_times)

print(f"Mean: {mean}, Median: {median}, Mode: {mode.mode[0]}")
```

# Measure of Dispersion
### Definition
Measures of dispersion describe the spread or variability in a data set. The main measures are range, variance, and standard deviation.

#### Example
Analyzing the variation in daily commute times

```python
commute_times = [30, 35, 32, 40, 28, 45, 50, 30, 33, 37]

# print(commute_times)

max_value = max(commute_times)
min_value = min(commute_times)

range_ans = max_value - min_value

print(range_ans)
```

# Python Example

```python
d = [30, 35, 32, 40, 28, 45, 50, 30, 33, 37]

range_ = np.ptp(d)
variance = np.var(d)
std_dev = np.std(d)

print(range_)
print(variance)
print(std_dev)
print(np.mean(d))
```

# Skewness and Kurtosis
### Definition
##### Skewness: 
Measures the asymmetry of the probability distribution of a real-valued random variable. A distribution can be positively skewed (right-skewed) or negatively skewed (left-skewed).


##### Kurtosis: 
Measures the "tailedness" of the probability distribution. High kurtosis indicates heavy tails, and low kurtosis indicates light tails.

#### Example
Analyzing income distribution in a community.

# Python Example

```python
income = [2000, 2200, 2500, 2700, 3000, 
          3500, 4000, 4500, 5000, 10000]


skewness = stats.skew(income)
kurtosis = stats.kurtosis(income)

print(f"""Skewness: {skewness}, 
      Kurtosis: {kurtosis}""")
```

# Covariance and Correlation
### Definition
##### Covariance:
Measures the directional relationship between two random variables. If both variables tend to increase or decrease together, the covariance is positive. If one variable tends to increase when the other decreases, the covariance is negative.


##### Correlation: 
Measures the strength and direction of the linear relationship between two variables. It is a standardized form of covariance and ranges between -1 and 1.

#### Example
Examining the relationship between hours of exercise and weight loss.

# Python Example

```python
exercise_hours = [1, 2, 3, 4, 5]
weight_loss = [2, 3, 5, 7, 9]

cov = \
np.cov(exercise_hours, weight_loss)[0][1]

corr = \
np.corrcoef(exercise_hours, weight_loss)[0][1]
print(cov)
print(corr)
```
